```python
import pandas as pd

data = {
    "product" : ["Laptop","Phone","Shoes","Watch","Laptop","Phone","Shoes","Watch",
                "Laptop","Phone","Shoes","Watch","Laptop","Phone","Shoes","Watch"],
    "sales" : [50000,20000,3000,5000,40000,25000,3500,6000,45000,22000,4000,5500,48000,27000,3200,7000],
    "regions" :["North","South","West","East","North","South","West","East",
                "North","South","West","East","North","South","West","East"]
}

df = pd.DataFrame(data)
df.head()
    
```

```python
df.info()
df.describe()
```

```python
df["sales"].sum()
```

```python
df.groupby("product")["sales"].sum()
```

```python
import matplotlib.pyplot as plt

df.groupby("product")["sales"].sum().plot(kind = "bar")
plt.title("product-wise sales")
plt.xlabel("product")
plt.ylabel("sales")
plt.show()
```

```python
df.groupby("regions")["sales"].sum().plot(kind="pie", autopct="%1.1f%%")
plt.title("regions-wise Sales")
plt.show()
```

#Insights

1. Laptop generates highest sales  
2. Shoes have lowest revenue  
3. North region has highest contribution
